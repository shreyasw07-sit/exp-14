```python
import pandas as pd
import numpy as np
data = {
  'Product':['laptop','mobile','tablet','headphone','smartwatch','camera'],
  'Price':[55000,20000,30000,2000,8000,45000],
  'unit_sold':[25,60,35,80,50,15]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Data Binning
bins1 = [0,10000,30000,60000]
labels1 = ['low','medium','high']
df['price_category'] = pd.cut(df['Price'],bins=bins1,labels=labels1)
print(df)
```

```python
#binning units sold
bins2_labels = ['low sales','medium sales','high sales']
df['sales_category'] = pd.cut(df['unit_sold'],bins=[0, 30, 65, 100], labels=bins2_labels)
print(df)
```

```python
# Data Formatting
#Convert Data Type
print(df.dtypes)
df['unit_sold'] = df['unit_sold'].astype(float)
print(df.dtypes)
```

```python
#Convert Text to Uppercase
df['Product'].str.upper()
df
```

```python
#Rounding Values
df['Price'] = df['Price'].round(2)
print(df)
```

```python
#Additional operation
#Sort Data
df.sort_values(by='Price',inplace=True)
df
```

```python
df.sort_values(by ='Price',inplace = True , ascending = False)
df
#decending order
```

```python
import pandas as pd
import numpy as np
data = {
    'Order_ID':['01','02','03','04','05','06','07','08'],
    'Order_Value' : [250,450, 120, 600, 300, 150, 500, 220],
    'Delivery_Time' : [30,45,25, 60, 35, 20, 50, 28],
     'Distance_km' : [5,8,3,10,6,2,9,4]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Binning Order Values
bins = [0,200,400,700]
labels = ['LOW','MEDIUM','HIGH']
df['Order_Category'] = pd.cut(df['Order_Value'],bins=bins,labels=labels)
print(df)
```

```python
#Binning Delivery Time
bins = [0,30,45,70]
labels2 = ['fast','normal','slow']
df['Delivery_speed'] = pd.cut(df['Delivery_Time'],bins=bins,labels=labels2)
print(df)
```

```python
#Binning Distance_km
bins_distance = [0, 4, 7, 10]
labels_distance = ['short', 'medium', 'long']
df['Distance_Category'] = pd.cut(df['Distance_km'], bins=bins_distance, labels=labels_distance)
print(df)
```

```python
# Sort DataFrame by 'Order_Value' in ascending order
print("Sorting by Order_Value (Ascending):")
print(df.sort_values(by='Order_Value', ascending=True))

```

```python
# Convert 'Order_Category' to uppercase
df['Order_Category_upper'] = df['Order_Category'].str.upper()
print("\nDataFrame with 'Order_Category' in uppercase:")
print(df[['Order_Category', 'Order_Category_upper']])
```

```python
# Data formating:
# Convert Data type

print(df.dtypes)
df['Distance_km'] = df['Distance_km'].astype(float)
print(df.dtypes)
```
